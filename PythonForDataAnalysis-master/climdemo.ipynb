```python
import pandas as pd
import sqlite3
import os


db = sqlite3.connect("LAC.db")

df1 = pd.read_csv("weather V3 02242020.csv", header = "infer")

df1.to_sql("weather", db, if_exists="replace")



def run_query(query):
    return pd.read_sql_query(query,db)

query = """
SELECT *
FROM weather
"""
run_query(query)
```

```python
query = """
SELECT substr("DATE", 1,10) AS "DATE",
substr("BackupDirection", 1,12) AS "BU Direction",
substr("HourlyDewPointTemperature", 1,25) AS "Hrly DP Temp",
substr("HourlyPrecipitation", 1,19) AS "Hrly Precip",
substr("HourlyPressureChange", 1,20) AS "Hrly Pressure Change",
substr("HourlyRelativeHumidity", 1,22) AS "Hrly Rel Humidity",
substr("HourlyWindSpeed", 1,15) AS "Hrly Wind Spd"
FROM weather;
"""
run_query(query)
```

```python
query = """
SELECT SUM("HourlyPrecipitation") AS total_precipitation 
FROM weather
"""
run_query(query)
```

```python
query = """
SELECT AVG("HourlyPrecipitation") AS avg_precipitation 
FROM weather
"""
run_query(query)
```
